```python
#Neccessary imports
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
from prophet.plot import plot_plotly
```

```python
#Loading of data
data=pd.read_csv("C:\\Users\\user\\Downloads\\investment-growth-forcast.csv")
```

```python
data.columns
```

```python
#Preparing the data
data['Depost']=pd.to_datetime(data['Deposit'])

#Rename of columns
data=data.rename(columns={'Deposit':'ds','Investment':'y'})
```

```python
#Initialize the model
model =Prophet()
```

```python
#Fit the model
model.fit(data)
```

```python
#Createing a dataframe for next 12 months
future=model.make_future_dataframe(periods=12, freq= 'M')
```

```python
#Make predictions
forecast=model.predict(future)
```

```python
#Plot the results
fig=model.plot(forecast)
plt.title('Savings Growth Forecast')
plt.xlabel('Date')
plt.ylabel('Growth')
plt.show()
```
